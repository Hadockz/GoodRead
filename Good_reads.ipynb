# Quais os 10 livros com mais avaliações?

```python
most_rated = df.sort_values('ratings_count', ascending=False)
most_rated.head(10)
```

```python
import seaborn as sns

sns.barplot(x=most_rated.head(10).ratings_count, y=most_rated.head(10).title)
```
